```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv("Sagittarius A.csv")
df2 = pd.read_csv("ori nebula.csv")
df3 = pd.read_csv("All_star.csv")
df4 = pd.read_csv("Horsehead_nebula.csv")

# Dividir por archivos
L1 = df1["phot_g_mean_mag"]
T1 = df1["teff_gspphot"]
C1 = df1["bp_rp"]

L2 = df2["phot_g_mean_mag"]
T2 = df2["teff_gspphot"]
C2 = df2["bp_rp"]

L3 = df3["phot_g_mean_mag"]
T3 = df3["teff_gspphot"]
C3 = df3["bp_rp"]

L4 = df4["phot_g_mean_mag"]
T4 = df4["teff_gspphot"]
C4 = df4["bp_rp"]

# Graficar los tres conjuntos de datos en el mismo gráfico con diferentes colores
plt.plot(C1, L1, "b+", label="Dataset 1")  # Azul
plt.plot(C2, L2, "r+", label="Dataset 2")  # Rojo
plt.plot(C3, L3, "g+", label="Dataset 3")  # Verde
plt.plot(C4, L4, "y+", label="Dataset 4")  # Amarillo

plt.grid()
plt.xlabel("Colour Index (BP-RP)")
plt.xlim(-0.5, 2.5)
plt.ylabel("Absolute Magnitude (G)")
plt.gca().invert_yaxis()
plt.title("Hertzsprung-Russell Diagram")
plt.legend()
plt.show()
```
