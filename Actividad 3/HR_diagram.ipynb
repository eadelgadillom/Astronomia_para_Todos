```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Eta_Carinae.csv")


L = df['phot_g_mean_mag']
T = df['teff_gspphot']
C = df['bp_rp']

plt.plot(T,L,"b+")
plt.grid()
plt.xlabel("Effective Temperature (K)")
plt.gca().invert_xaxis()
plt.ylabel("Absolute Magnitude (G)")
plt.gca().invert_yaxis()
plt.title("Hertzsprung-Russell Diagram")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv("Betelgeuse.csv")
df2 = pd.read_csv("Eta_Carinae.csv")
df3 = pd.read_csv("Sirius_B.csv")
df4 = pd.read_csv("IK Pegasi B.csv")

# Dividir por archivos
L1 = df1["phot_g_mean_mag"]
T1 = df1["teff_gspphot"]
C1 = df1["bp_rp"]

L2 = df2["phot_g_mean_mag"]
T2 = df2["teff_gspphot"]
C2 = df2["bp_rp"]

L3 = df3["phot_g_mean_mag"]
T3 = df3["teff_gspphot"]
C3 = df3["bp_rp"]

L4 = df4["phot_g_mean_mag"]
T4 = df4["teff_gspphot"]
C4 = df4["bp_rp"]

# Graficar los tres conjuntos de datos en el mismo grfico con diferentes colores
plt.plot(T1, L1, "r+", label="Dataset 1")  # Azul
plt.plot(T2, L2, "b+", label="Dataset 2")  # Rojo
plt.plot(T3, L3, "g+", label="Dataset 3")  # Verde
plt.plot(T4, L4, "y+", label="Dataset 4")  # Amarillo

plt.grid()
plt.xlabel("Effective Temperature (K)")
plt.gca().invert_xaxis()
plt.ylabel("Absolute Magnitude (G)")
plt.gca().invert_yaxis()
plt.title("Hertzsprung-Russell Diagram")
plt.legend()
plt.show()
```
