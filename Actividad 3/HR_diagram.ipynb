## DIAGRAMA HR (Hertzsprung - Russell)



Creacin con catalogo Gaia



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Eta_Carinae.csv")


L = df['phot_g_mean_mag']
T = df['teff_gspphot']
C = df['bp_rp']

plt.plot(T,L,"b+")
plt.grid()
plt.xlabel("Effective Temperature (K)")
plt.gca().invert_xaxis()
plt.ylabel("Absolute Magnitude (G)")
plt.gca().invert_yaxis()
plt.title("Hertzsprung-Russell Diagram")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv("Betelgeuse.csv")
df2 = pd.read_csv("Eta_Carinae.csv")
df3 = pd.read_csv("Sirius_B.csv")
df4 = pd.read_csv("IK Pegasi B.csv")

# Dividir por archivos
L1 = df1["phot_g_mean_mag"]
T1 = df1["teff_gspphot"]
C1 = df1["bp_rp"]

L2 = df2["phot_g_mean_mag"]
T2 = df2["teff_gspphot"]
C2 = df2["bp_rp"]

L3 = df3["phot_g_mean_mag"]
T3 = df3["teff_gspphot"]
C3 = df3["bp_rp"]

L4 = df4["phot_g_mean_mag"]
T4 = df4["teff_gspphot"]
C4 = df4["bp_rp"]

# Graficar los tres conjuntos de datos en el mismo grfico con diferentes colores
plt.plot(T1, L1, "r+", label="Dataset 1")  # Azul
plt.plot(T2, L2, "b+", label="Dataset 2")  # Rojo
plt.plot(T3, L3, "g+", label="Dataset 3")  # Verde
plt.plot(T4, L4, "y+", label="Dataset 4")  # Amarillo


plt.grid()
plt.xlabel("Effective Temperature (K)")
plt.gca().invert_xaxis()
plt.ylabel("Absolute Magnitude (G)")
plt.gca().invert_yaxis()
plt.title("Hertzsprung-Russell Diagram")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

archivos = ["Betelgeuse.csv", "Eta_Carinae.csv", "Sirius_B.csv", "IK Pegasi B.csv"]
colores = ["r", "b", "g", "y"]

plt.figure(figsize=(8, 6))

for archivo, color in zip(archivos, colores):
    df = pd.read_csv(archivo)
    temperatura = df["teff_gspphot"]
    magnitud = df["phot_g_mean_mag"] 
    plt.plot(temperatura, magnitud, color + "o", label=archivo)

# ----------- Lneas de referencia del diagrama H-R -----------

# Temperatura (K) y magnitud absoluta en Gaia G para la secuencia principal
temp_main_seq = [30000, 20000, 10000, 7500, 6000, 5000, 4000, 3000,2500]
mag_main_seq = [0,1, 4, 6, 8, 9, 10, 12, 15]

# Supergigantes
temp_supergiants = [25000, 15000, 7000, 5000, 4000]
mag_supergiants = [-8, -7, -5, -3, -2]

# Enanas blancas
temp_white_dwarfs = [25000, 20000, 15000, 10000]
mag_white_dwarfs = [10, 12, 14, 15]

# Graficar lneas de referencia
plt.plot(temp_main_seq, mag_main_seq, "m-", label="Main Sequence")  # Secuencia principal (violeta)
plt.plot(temp_supergiants, mag_supergiants, "r-", label="Supergiants")  # Supergigantes (rojo)
plt.plot(temp_white_dwarfs, mag_white_dwarfs, "c-", label="White Dwarfs")  # Enanas blancas (cyan)

#plt.xscale("log")  # Escala logartmica en temperatura
plt.gca().invert_xaxis()  
plt.gca().invert_yaxis()  
plt.grid()
plt.xlabel("Surface Temperature (K)")
plt.ylabel("Gaia G Absolute Magnitude")
plt.title("Hertzsprung-Russell Diagram (Gaia)")
plt.legend()
plt.show()
```
